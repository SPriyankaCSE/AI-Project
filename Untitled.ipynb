```python
from collections import deque

goal_state = [1, 2, 3, 4, 5, 6, 7, 8, 0]  # 0 represents the blank

def bfs(start_state):
    visited = set()
    queue = deque([(start_state, [])])

    while queue:
        state, path = queue.popleft()
        if state == goal_state:
            return path + [state]

        visited.add(tuple(state))
        for neighbor in get_neighbors(state):
            if tuple(neighbor) not in visited:
                queue.append((neighbor, path + [state]))

    return None

def get_neighbors(state):
    neighbors = []
    i = state.index(0)  # index of blank
    row, col = divmod(i, 3)
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up Down Left Right

    for dr, dc in moves:
        r, c = row + dr, col + dc
        if 0 <= r < 3 and 0 <= c < 3:
            j = r * 3 + c
            new_state = state[:]
            new_state[i], new_state[j] = new_state[j], new_state[i]
            neighbors.append(new_state)

    return neighbors

# Example input
start = [1, 2, 3, 4, 0, 6, 7, 5, 8]
solution = bfs(start)

if solution:
    print("BFS solution:")
    for s in solution:
        print(s[:3])
        print(s[3:6])
        print(s[6:])
        print()
else:
    print("No solution found.")
```

```python
from sklearn.datasets import load_iris
from heapq import heappop, heappush
import numpy as np

# Load Iris dataset from UCI
iris = load_iris()
data = iris.data
target = iris.target

# Heuristic: Euclidean distance to goal point
def heuristic(a, b):
    return np.linalg.norm(a - b)

def a_star_search(start_index, goal_index):
    start = data[start_index]
    goal = data[goal_index]

    open_set = []
    heappush(open_set, (0 + heuristic(start, goal), 0, start_index, []))  # (f, g, index, path)
    visited = set()

    while open_set:
        f, g, current_index, path = heappop(open_set)
        if current_index == goal_index:
            return path + [current_index]

        if current_index in visited:
            continue

        visited.add(current_index)

        for i in range(len(data)):
            if i not in visited:
                cost = np.linalg.norm(data[current_index] - data[i])  # move cost
                h = heuristic(data[i], goal)
                heappush(open_set, (g + cost + h, g + cost, i, path + [current_index]))

    return None

# Test A* on Iris dataset
start_idx = 0   # first sample
goal_idx = 149  # last sample
path = a_star_search(start_idx, goal_idx)

print("A* Path from sample", start_idx, "to", goal_idx)
for i in path:
    print(f"Index: {i}, Features: {data[i]}, Class: {target[i]}")
```

```python
import heapq

# Sample state graph where node names are concepts/images
graph = {
    "Start": [("Sports", 2)],
    "Sports": [("Ball Sports", 2)],
    "Ball Sports": [("Basketball", 1), ("Football", 3)],
    "Basketball": [("Basketball Logo", 1)],
    "Football": [],
    "Basketball Logo": []
}

# Heuristic values for nodes (estimated cost to goal)
heuristics = {
    "Start": 5,
    "Sports": 4,
    "Ball Sports": 3,
    "Basketball": 1,
    "Football": 5,
    "Basketball Logo": 0
}

# SMA* implementation with memory bound
def sma_star(start, goal, memory_limit=5):
    frontier = []
    heapq.heappush(frontier, (heuristics[start], 0, start, []))  # (f, g, current_node, path)

    while frontier:
        # Enforce memory bound
        if len(frontier) > memory_limit:
            frontier.sort(reverse=True)  # sort descending f
            frontier.pop()  # remove worst node

        f, g, current, path = heapq.heappop(frontier)
        new_path = path + [current]

        if current == goal:
            return new_path

        for neighbor, cost in graph.get(current, []):
            g_new = g + cost
            f_new = g_new + heuristics[neighbor]
            heapq.heappush(frontier, (f_new, g_new, neighbor, new_path))

    return None

# Run the search
path = sma_star("Start", "Basketball Logo", memory_limit=3)

# Output
if path:
    print("Path to Basketball Logo:")
    for step in path:
        print("->", step)
else:
    print("Logo not found within memory constraints.")
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import random
import networkx as nx
import matplotlib.pyplot as plt

# Function to generate a random Bayesian Network structure
def generate_random_bayesian_network(num_nodes, edge_probability):
    graph = nx.erdos_renyi_graph(num_nodes, edge_probability, directed=True)
    # Ensure the graph is acyclic
    while not nx.is_directed_acyclic_graph(graph):
        graph = nx.erdos_renyi_graph(num_nodes, edge_probability, directed=True)
    mapping = {i: f'node_{i}' for i in range(num_nodes)}
    graph = nx.relabel_nodes(graph, mapping)
    return BayesianNetwork(graph.edges())

# Function to define random CPDs for the network
def define_random_cpds(model):
    for node in model.nodes():
        parents = list(model.predecessors(node))
        if parents:
            # Random CPD for nodes with parents
            data = [[random.random() for _ in range(2**len(parents))] for _ in range(2)]
            for i in range(len(data)):
                s = sum(data[i])
                data[i] = [v/s for v in data[i]]
            cpd = TabularCPD(node, 2, data, evidence=parents, evidence_card=[2]*len(parents))
        else:
            # Random CPD for root nodes
            prob = random.random()
            cpd = TabularCPD(node, 2, [[prob], [1-prob]])
        model.add_cpds(cpd)

# Function to perform inference on the network
def perform_inference(model, query_node, evidence=None):
    inference = VariableElimination(model)
    result = inference.query(variables=[query_node], evidence=evidence)
    return result

# Function to run iterations and analyze results
def run_iterations(num_iterations, num_nodes, edge_probability, query_node):
    results = []
    for _ in range(num_iterations):
        model = generate_random_bayesian_network(num_nodes, edge_probability)
        define_random_cpds(model)

        # Provide evidence for all other nodes except the query node
        evidence = {node: random.randint(0, 1) for node in model.nodes() if node != query_node}

        result = perform_inference(model, query_node, evidence=evidence)
        results.append(result.values[1]) # Probability of query node being True

    return results

# Main execution
num_iterations = 100
num_nodes = 5
edge_probability = 0.5
query_node = 'node_0'

results = run_iterations(num_iterations, num_nodes, edge_probability, query_node)

# Analyze the results
average_probability = sum(results) / len(results)
print(f"Average probability of {query_node} being True: {average_probability}")

# Visualize the distribution of probabilities
plt.hist(results, bins=20)
plt.xlabel(f"Probability of {query_node} being True")
plt.ylabel("Frequency")
plt.title(f"Distribution of Probabilities for {query_node}")
plt.show()
```
